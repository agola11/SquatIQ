# Squat Sensor Experimentation

## Imports

```python
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Configure matplotlib to show figures embedded in the notebook
%matplotlib inline

NUM_SENSORS = 5
```

## Utilities

```python
# data: an N by num_sensors matrix, where N number of readings
# returns a list of length num_sensors, each is an array of shape (N,)
# h: smoothing filter
def preprocess_sensor_data(data, num_sensors, h):
    return np.apply_along_axis(lambda row: np.convolve(row, h, mode='valid'), axis=0, arr=data)
```

## Example: read in sensor data

```python
sensor_data = np.genfromtxt('input_heel_lift.csv', delimiter=',')
N, M = sensor_data.shape

sensor_data = sensor_data[0:-1]

plt.figure()
plt.title("before smoothing")
plt.plot(sensor_data)
plt.show()

h = np.ones(10)
h = h / np.sum(h)

sensor_data_smth = preprocess_sensor_data(sensor_data, NUM_SENSORS, h)

plt.figure()
plt.title("after smoothing")
plt.plot(sensor_data_smth)
plt.show()
```

## Read in actual sensor data to be used for training

```python
sensr_dir = 'raw_sensor_data/'

input_files = {'hlpron/hlpron_1.csv': (450, 1100, 'hlpron'), 
               'hlsup/hlsup_1.csv': (200, 700, 'hlsup'), 
               'sup/sup_1.csv': (300, 1400, 'sup'),
               'heellift/heellift_1.csv': (300, 1100, 'heellift'),
               'pron/pron_1.csv': (200, 1100, 'pron')}

sensor_data_train = {}

for fn in input_files.keys():
    sensor_data = np.genfromtxt(sensr_dir + fn, delimiter=',')
    sensor_data = sensor_data[input_files[fn][0]:input_files[fn][1]]
    #sensor_data_smth = preprocess_sensor_data(sensor_data, NUM_SENSORS, h)
    sensor_data_train[input_files[fn][2]] = sensor_data
    plt.figure()
    plt.title(fn)
    plt.plot(sensor_data)
    plt.show()
```

## Classifying Utilities 

```python
classes = {'hlpron': 0, 
           'hlsup': 1, 
           'sup': 2,
           'heellift': 3,
           'pron': 4}

classes_inv = {0:'hlpron', 
               1:'hlsup', 
               2:'sup',
               3:'heellift',
               4:'pron'}

# expects array-like data
def create_training_data(data, offset=5):
    N, M = data.shape
    idx = 0
    X_t = []
    while idx + offset <= N:
        X_t.append(data[idx:idx+offset].flatten(order='F'))
        idx += offset
    return np.array(X_t)

test_sensor_data = np.genfromtxt('input_heel_lift.csv', delimiter=',')
N, M = test_sensor_data.shape

test_sensor_data = test_sensor_data[200:1200]


plt.figure()
plt.title("test_1")
plt.plot(test_sensor_data)
plt.show()

X_test = create_training_data(test_sensor_data)

print(X_test.shape)
```

## Train a multiclass classifier

```python
Xs = []
Ys = []
for key in sensor_data_train.keys():
    data = sensor_data_train[key]
    X_t = create_training_data(data)
    N_t, M_t = X_t.shape
    Xs.append(X_t)
    Ys.append([classes[key]] * N_t)

X = np.vstack(Xs)
Y = np.hstack(Ys)

print (X.shape)
print (Y.shape)

from sklearn.svm import LinearSVC

multi_clf = LinearSVC().fit(X, Y)
```

## Test Multiclass Model

```python
Z = multi_clf.predict(X_test)

Z_str = [classes_inv[val] for val in Z]

print (Z)

plt.figure()
plt.title("after smoothing")
plt.plot(Z)
plt.show()
```

## Save model

```python
from sklearn.externals import joblib

joblib.dump(multi_clf, 'clf_v1.pkl') 
```
